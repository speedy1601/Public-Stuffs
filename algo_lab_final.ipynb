# Binary Searching and Insertion Sort

```python
def binarySearch(arr: list, item: int):
    if len(arr) == 0:
        return "The array is empty. Insert some element please!"
    
    n, arr1 = len(arr), arr.copy()
    if arr_not_sorted(arr):
        arr1 = insertionSort(arr)

    low, high, mid = 0, len(arr1) - 1, 0

    while low < high:
        mid = (low + high) >> 1
        if arr1[mid] == item:
            break
        elif arr1[mid] < item:
            low = mid + 1
        else:
            high = mid - 1
    
    return "The number is not in the array" if arr1[mid] != item else f"The index of value {item} in original array = {[i for i in range(n) if arr[i] == item][0]} and in sorted array = {mid}."

def arr_not_sorted(arr):
    for i in range(1, len(arr)):
        if arr[i] < arr[i-1]: # if next element is < previous element, the array is unsorted.
            return True
            
    print("The array is already sorted. No need to sort it to perform binary search.")
    return False

def insertionSort(arr):
    arr1 = arr.copy()
    for i in range(1, len(arr1)):
        cur_value = arr1[i]
        if arr1[i] < arr1[i-1]: # In insertionSort, the previous elements are already sorted. If cur_value(5) is < prev_value(4)
                                # , that means we dont need to perform sorting as its already sorted. If not, then sort it.
            for j in range(i-1, -1, -1): # This loop is to place the cur_value in its sorted position if above "if" statement is true.
                if cur_value >= arr1[j]: # After continuing to place cur_value in correct position, if cur_value(next value) is >= prev_value(j), break.
                    break
                arr1[j], arr1[j+1] = arr1[j+1], arr1[j] # 1, 3, 4(j), 2(j+1) -> 1, 3(j), 2(j+1), 4 -> 1(j), 2(j+1), 3, 4
    
    print("Array need to be sorted to perform BS which is ", arr1)
    return arr1

n = int(input("Enter the size of the array : "))
arr = [int(v) for v in input(f"Enter {n} numbers : ").split()][:n]
print("The original array is", arr)

search = int(input("Enter the number whose index you want to search : "))
print(binarySearch(arr, search))
```

# Merge Sort

```python
def merge_both_left_and_right_sorted_subarray(arr, left, mid, right):
    # We break the arr into 2 array, left_subarray and right_subarray. Then sort the arr IN-PLACE by traversing those 2 arrays.
    
    left_subarray_size, right_subarray_size = (mid - left + 1), (right - mid)

    left_subarray  = [None] * left_subarray_size
    right_subarray = [None] * right_subarray_size

    for l in range(left_subarray_size):
        left_subarray[l] = arr[left + l]

    for r in range(right_subarray_size):
        right_subarray[r] = arr[mid + 1 + r]
    
    l, r, i = 0, 0, left

    while l < left_subarray_size and r < right_subarray_size:
        if left_subarray[l] <= right_subarray[r]:
            arr[i] = left_subarray[l]
            l += 1
        else:
            arr[i] = right_subarray[r]
            r += 1
        i += 1
    
    if l < left_subarray_size:
        while l < left_subarray_size:
            arr[i] = left_subarray[l]
            l += 1
            i += 1
    elif r < right_subarray_size:
        while r < right_subarray_size:
            arr[i] = right_subarray[r]
            r += 1
            i += 1

def mergeSort(arr, left, right):
    if left < right:
        mid = (left + right) >> 1
        mergeSort(arr, left, mid)
        mergeSort(arr, mid + 1, right)
        merge_both_left_and_right_sorted_subarray(arr, left, mid, right)


def main():
    n = int(input("Enter the size of the array : "))

    arr = [int(v) for v in input(f"Enter {n} numbers : ").split(' ')] [:n]
    # print(arr, len(arr), type(arr[0]))
    print("The original array is :", arr)

    mergeSort(arr, 0, n - 1)

    print("The sorted array is :", arr)

if __name__ == "__main__":
    main()
```

```python
# Summary : We're trying including the current item in the bag by "if" black. 
# Take user input for weights and profits. Then sort the weights and based on sorted weight, the profits would sort too.

def insertionSort(arr1, arr2):
    for i in range(1, len(arr1)):
        cur_value = arr1[i]
        if arr1[i] < arr1[i-1]: # In insertionSort, the previous elements are already sorted. If cur_value(5) is < prev_value(4)
                                # , that means we dont need to perform sorting as its already sorted. If not, then sort it.
            for j in range(i-1, -1, -1): # This loop is to place the cur_value in its sorted position if above "if" statement is true.
                if cur_value >= arr1[j]: # After continuing to place cur_value in correct position, if cur_value(next value) is >= prev_value(j), break.
                    break
                arr1[j], arr1[j+1] = arr1[j+1], arr1[j] # 1, 3, 4(j), 2(j+1) -> 1, 3(j), 2(j+1), 4 -> 1(j), 2(j+1), 3, 4
                arr2[j], arr2[j+1] = arr2[j+1], arr2[j]

def merge_both_left_and_right_sorted_subarray(arr, left, mid, right, arr1):
    # We break the arr into 2 array, left_subarray and right_subarray. Then sort the arr IN-PLACE by traversing those 2 arrays.
    
    left_subarray_size, right_subarray_size = (mid - left + 1), (right - mid)

    left_subarray  = [None] * left_subarray_size
    left_subarray1 = left_subarray.copy()
    right_subarray = [None] * right_subarray_size
    right_subarray1 = right_subarray.copy()

    for l in range(left_subarray_size):
        left_subarray[l]  = arr[left + l]
        left_subarray1[l] = arr1[left + l]

    for r in range(right_subarray_size):
        right_subarray[r]  = arr[mid + 1 + r]
        right_subarray1[r] = arr1[mid + 1 + r]
    
    l, r, i = 0, 0, left

    while l < left_subarray_size and r < right_subarray_size:
        if left_subarray[l] <= right_subarray[r]:
            arr[i]  = left_subarray[l]
            arr1[i] = left_subarray1[l]
            l += 1
        else:
            arr[i]  = right_subarray[r]
            arr1[i] = right_subarray1[r]
            r += 1
        i += 1
    
    if l < left_subarray_size:
        while l < left_subarray_size:
            arr[i]  = left_subarray[l]
            arr1[i] = left_subarray1[l]
            l += 1
            i += 1
    elif r < right_subarray_size:
        while r < right_subarray_size:
            arr[i]  = right_subarray[r]
            arr1[i] = right_subarray1[r]
            r += 1
            i += 1

def mergeSort(arr, left, right, arr1):
    if left < right:
        mid = (left + right) >> 1
        mergeSort(arr, left,    mid,   arr1)
        mergeSort(arr, mid + 1, right, arr1)
        merge_both_left_and_right_sorted_subarray(arr, left, mid, right, arr1)

def arr_not_sorted(arr):
    for i in range(1, len(arr)):
        if arr[i] < arr[i-1]: # if next element is < previous element, the array is unsorted.
            return True
            
    print("The array is already sorted. No need to sort it to perform binary search.")
    return False

def knapsack(capacity, profits, weights):
    weights_of = weights.copy()
    profits_of = profits.copy()

    if arr_not_sorted(weights_of):
        mergeSort(weights_of, 0, len(profits_of) - 1, profits_of) # merge sort.
        # insertionSort(weights_of, profits_of) # insertion sort.
    
    print("\nSorted Weights :", weights_of)
    print("Sorted Profits :", profits_of)
    
    total_items = len(weights_of)
    dp = [[0] * (capacity + 1) for _ in range(total_items + 1)]

    for cur_item_in_dp in range(1, total_items + 1):
        for cur_capacity in range(1, capacity + 1):
            # both "cur_item_in_bag" and "cur_item_in_dp" as in "index" refers to the same item but in bag and dp respectively.
            cur_item_in_bag, prev_item_in_dp = cur_item_in_dp - 1, cur_item_in_dp - 1
            remaining_space_after_taking_current_item = cur_capacity - weights_of[cur_item_in_bag]

            if weights_of[cur_item_in_bag] <= cur_capacity:
                dp[cur_item_in_dp][cur_capacity] = max(dp[prev_item_in_dp][cur_capacity], 
                                                       profits_of[cur_item_in_bag] + dp[prev_item_in_dp][remaining_space_after_taking_current_item])
            else : # if current capacity itself is less than the weight of the current item, then we can't include that item.
                dp[cur_item_in_dp][cur_capacity] = dp[prev_item_in_dp][cur_capacity]

    return dp[total_items][capacity]

if __name__ == "__main__":
    n = int(input("Enter the size of the array : "))
    
    weights  = [int(v) for v in input(f"Enter {n} numbers for 'weights': ").split(' ')] [:n] # [2 5  1 3]
    profits  = [int(v) for v in input(f"Enter {n} numbers for 'profits': ").split(' ')] [:n] # [6 16 1 10]
    capacity = [int(v) for v in input(f"Enter the value of capacity    : ").split(' ')] [0]  # 7. Answer is 22.
    
    print("\nOriginal Weights :", weights)
    print("Original Profits :", profits)

    print(knapsack(capacity, profits, weights))

"""
if we can include the current item in the bag:
    max profit for current item in current capacity = max(previous max profit exluding the current item in case including the current item doesn't yield max profit,
                                                          max profit can be yielded including the current item)
                                                          
else : # if we can't include the current item in bag, then the profit is simply it;s previous item's profit.
    max profit for current item in current capacity = the previous item's max profit for the current capacity since we can't include the current item in the bag at all.

max(option1, option2)

    Option 1 : Choose the previous result if including the current item in the bag doesn't yield maximum profit.
    Option 2 : Include the current item by its profit + what the remaining space excluding the current item can yield MAX PROFIT in previous item where
           the current item is excluded.

"""
```
